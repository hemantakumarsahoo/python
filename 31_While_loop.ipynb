 A program to display a pattern using loops
 n=6
 .....*                 i   .   *
 ....***               -------------
 ...*****               1   5    1
 ..*******              2   4    3  
 .*********             3   3    5
 ***********            4   2    7
                        5   1    9
                        6   0    11

logic:
.=n-i
*=2*i-1

```python
n=int(input("Enter the number of layers ...."))
i=1
while i<=n:
    j=1
    while  j<= n-i:
        print(" ", end="")
        j+=1
    j=1
    while j<=2*i-1:
        print("*",end="")
        j+=1
    print()
    i+=1
```
